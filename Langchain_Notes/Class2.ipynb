```python
import os

from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(api_key=os.getenv("OPENAI_API_KEY"),
                  base_url="https://api.longcat.chat/openai",
                  model="LongCat-Flash-Chat",
                  max_tokens=1000,
                  temperature=0)
```

```python
from langchain_classic.memory import ConversationBufferMemory
memory = ConversationBufferMemory(memory_key="history")
```

```python
from langchain_classic.chains import ConversationChain
```

```python
conversation = ConversationChain(
    llm=llm, 
    memory=memory,
    verbose=False
    )
```

```python
conversation.predict(input = "hello, my name is Veinsure")
```

```python
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)
```

```python
conversation.predict(input="hello, my name is Veinsure")
```

```python
print(memory.buffer)
```

```python
memory.load_memory_variables({})
```

```python
memory = ConversationBufferMemory()
```

```python
memory.save_context({"input": "hello, my name is Veinsure"}, {"output": "hello, my name is Veinsure"})
memory.load_memory_variables({})
memory.save_context({"input": "I'm a student"}, {"output": "OK, I know."})
memory.load_memory_variables({})
memory.save_context({"input": "I like Python"}, {"output": "OK, I know."})
memory.load_memory_variables({})
```
