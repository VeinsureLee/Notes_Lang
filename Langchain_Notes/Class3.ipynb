````python
from langchain_classic.chains import SimpleSequentialChain, SequentialChain
````

````python
llm = ChatOpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url="https://api.longcat.chat/openai",
    model="LongCat-Flash-Chat",
    max_tokens=1000,
    temperature=0.9
)
````

````python
# Chain 1
first_prompts = ChatPromptTemplate.from_template(
    "给制作{product}的公司起一个名字"
)


first_chain = LLMChain(
    llm=llm, 
    prompt=first_prompts, 
    output_key="company_name", 
    verbose = True
    )
````

````python
# Chain 2
second_prompts = ChatPromptTemplate.from_template(
    "给{company_name}的公司写一个一句话的描述:"
)

second_chain = LLMChain(
    llm=llm, 
    prompt=second_prompts, 
    output_key="review",
    verbose = True
    )
````

````python
# Chain 3
third_prompts = ChatPromptTemplate.from_template(
    "{review}用的什么语言"
)

third_chain = LLMChain(
    llm=llm, 
    prompt=third_prompts, 
    output_key="language",
    verbose=True
    )
````

````python
# Chain 4
fourth_prompts = ChatPromptTemplate.from_template(
    "{language}的创始人是谁"
)

fourth_chain = LLMChain(
    llm=llm, 
    prompt=fourth_prompts, 
    output_key="founder",
    verbose=True
    )
````

````python
overall_simple_chain = SequentialChain(
    chains=[first_chain, second_chain, third_chain, fourth_chain], 
    input_variables=["product"],
    output_variables=["company_name", "review", "language", "founder"],
    verbose=True
    )
````

````python
result = overall_simple_chain.invoke({"product": "Banana Phone"})
````

下面举一个根据问题类型路由到不同位置的例子

````python
math_template = """你是一个数学专家，请你根据下面问题，从数学角度回答。
问题如下：
{input}
"""
computer_template = """你是一个计算机专家，请你根据下面问题，从计算机科学角度回答。
问题如下：
{input}
"""
````

````python
prompt_infos = [
    {
        "name": "数学",
        "description": "解决数学问题",
        "prompt_template": math_template
    },
    {
        "name": "计算机",
        "description": "解决计算机问题",
        "prompt_template": computer_template
    },
    {
        "name": "默认",
        "description": "数学和计算机以外的问题（如生物、历史、常识等），直接按通用方式回答",
        "prompt_template": "{input}"  # 默认链直接用用户问题，不加专家人设
    }
]
````

````python
from langchain_classic.chains.router import MultiPromptChain
from langchain_classic.chains.router.llm_router import LLMRouterChain, RouterOutputParser
from langchain_classic.prompts import PromptTemplate
````

````python
llm = ChatOpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url="https://api.longcat.chat/openai",
    model="LongCat-Flash-Chat",
    max_tokens=1000,
    temperature=0.9
)
````

````python
destination_chains = {}
for p_info in prompt_infos:
    name = p_info["name"]
    prompt_template = p_info["prompt_template"]
    prompt = ChatPromptTemplate.from_template(prompt_template)
    chain = LLMChain(llm=llm, prompt=prompt)
    destination_chains[name] = chain
````

````python
destinations = [
    f"{p_info['name']}: {p_info['description']}" for p_info in prompt_infos]
destinations_str = "\n".join(destinations)
````

````python
# 默认模版必须包含 {input} 占位符，否则用户问题不会传入 LLM
default_prompt = ChatPromptTemplate.from_template("{input}")
default_chain = LLMChain(llm=llm, prompt=default_prompt)
````

````python
# 注意：LangChain 的 parse_json_markdown 只识别反引号 ``` 的代码块，不能用 '''json
MULTI_PROMPT_ROUTER_TEMPLATE = """
给一个文本回答，格式如下：

<<<FORMATTING>>>
用 markdown 代码块包裹 JSON。注意：第一行必须是 ```json（三个反引号与 json 紧挨同一行，中间不能换行），否则会解析失败：
```json
{{{{
    "destination": string,
    "next_inputs": string
}}}}
```
REMEMBER: "destination" 必须是下面 CANDIDATE PROMPTS 中的一个名称（数学、计算机、默认）。若问题不属于数学或计算机领域，选「默认」。
REMEMBER: "next_inputs" 必须是字符串，即用户的问题原文（例如 "1+1=?"）

<< CANDIDATE PROMPTS >>
{destinations}

<< INPUT >>
{{input}}

<< OUTPUT（只输出一行 ```json 紧接着换行再写 JSON，最后一行 ```，不要先 ``` 再换行再写 json）>>
"""
````

````python
router_template = MULTI_PROMPT_ROUTER_TEMPLATE.format(
    destinations=destinations_str
    )
router_prompt = PromptTemplate(
    template=router_template,
    input_variables=["input"],
    output_parser=RouterOutputParser()
    )
router_chain = LLMRouterChain.from_llm(llm, router_prompt)
````

````python
chain = MultiPromptChain(
    router_chain=router_chain,
    destination_chains=destination_chains,
    default_chain=default_chain,
    verbose=True
    )
````

````python
chain.invoke({"input": "1+1=?"})
````

````python
chain.invoke({"input": "显卡是什么"})
````
